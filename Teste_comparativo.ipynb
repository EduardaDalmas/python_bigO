# Código de Compromissos

```python
import random
```

```python
import datetime
```

```python
class Compromisso:
  def __init__(self, data, hora, descricao):
    self.data = data
    self.hora = hora
    self.descricao = descricao

  def __lt__(self, other):
      return datetime.datetime.strptime(self.data + ' ' + self.hora, '%d/%m/%Y %H:%M') < datetime.datetime.strptime(other.data + ' ' + other.hora, '%d/%m/%Y %H:%M')

  def __str__(self):
    return f"{self.data} - {self.hora} - {self.descricao}"


class Agenda():
  def __init__(self, capacidade):
    self.capacidade = capacidade
    self.ultima_posicao = -1
    self.valores = [None] * self.capacidade

  def adicionar(self, valor):
    if self.ultima_posicao == self.capacidade -1:
      print('Capacidade máxima atingida')
      return 
    posicao = 0
    for i in range(self.ultima_posicao + 1):
      posicao = i
      if self.valores[i] > valor:
        break
      if i == self.ultima_posicao:
        posicao = i + 1
    x = self.ultima_posicao
    while x >= posicao:
      self.valores[x + 1] = self.valores[x]
      x -= 1
    self.valores[posicao] = valor
    self.ultima_posicao += 1 
    self.ordena_vetor(self.valores)

  def ordena_vetor(self, valores):
    valores_nao_vazios = filter(lambda valor: valor is not None, self.valores)
    ordenacao = sorted(valores_nao_vazios, key=lambda valor: valor.data, reverse=True)
    for i, valor in enumerate(ordenacao):
      print(f'indice {i} - Data: {valor.data} - hora: {valor.hora}, descricao: {valor.descricao}')

  def excluir(self, valor):
    if self.ultima_posicao == -1:
      print('Array vazio')
      return 
    else:
      for i in range(valor, self.ultima_posicao):
        self.valores[i] = self.valores[i + 1]
      self.ultima_posicao -= 1

  def buscar(self, valor, tipo):
    limite_inferior = 0
    limite_superior = self.ultima_posicao

    while True:
      posicao_atual = int((limite_inferior + limite_superior) / 2)
      
      if tipo == 'data':
        if self.valores[posicao_atual].data == valor:
          return posicao_atual

        elif limite_inferior > limite_superior:
          return -1
        
        else:
          if self.valores[posicao_atual].data < valor:
            limite_inferior = posicao_atual + 1
          else:
            limite_superior = posicao_atual - 1
      
      if tipo == 'hora':
        if self.valores[posicao_atual].hora == valor:
          return posicao_atual

        elif limite_inferior > limite_superior:
          return -1
        
        else:
          if self.valores[posicao_atual].hora < valor:
            limite_inferior = posicao_atual + 1
          else:
            limite_superior = posicao_atual - 1

      if tipo == 'descricao':
        if self.valores[posicao_atual].descricao == valor:
          return posicao_atual

        elif limite_inferior > limite_superior:
          return -1
        
        else:
          if self.valores[posicao_atual].descricao < valor:
            limite_inferior = posicao_atual + 1
          else:
            limite_superior = posicao_atual - 1      

  def visualizar(self):
    if self.ultima_posicao == -1:
      print('O vetor está vazio')   
    else:
      for i in range(self.ultima_posicao +1):
        compromisso = self.valores[i]
        print('posicao: ', i, ' | data: ', compromisso.data, ' | hora: ', compromisso.hora, ' | descricao: ', compromisso.descricao) 
```

2. Estoque de uma Loja Online:

Crie um catálogo de produtos em um vetor não ordenado, onde cada elemento contém informações sobre o produto, como nome, código, quantidade, preço e avaliação de clientes. 

a) Implemente um algoritmo para adicionar e excluir produtos no vetor não ordenado.

b) Implemente  um algoritmo de busca para permitir aos funcionários procurarem por produtos específicos por código, nome ou outra informação relevante.

c) Implemente um algoritmo de busca ou ordenação para encontrar os melhores produtos, por exemplo, aqueles com as melhores avaliações ou melhores preços. Retorne os produtos encontrados para exibição aos usuários.

d) Implemente também a visualização de todos os produtos do estoque e cálculo de total de estoque.

```python
class Produto:
  def __init__(self, nome, codigo, quantidade, preco, avaliacao):
    self.nome = nome
    self.codigo = codigo
    self.quantidade = quantidade
    self.preco = preco
    self.avaliacao = avaliacao

class Estoque():
  def __init__(self, capacidade):
    self.capacidade = capacidade
    self.ultima_posicao = -1
    self.produtos = [None] * self.capacidade

  def adicionar(self, valor):
    if self.ultima_posicao == self.capacidade -1:
      print('Capacidade máxima atingida')
    else:
      self.ultima_posicao  += 1
      self.produtos[self.ultima_posicao] = valor
  
  def buscar(self, valor, tipo):
    if tipo == 'codigo':
      for i in range(self.ultima_posicao + 1):
        codigo_produto = self.produtos[i]
        if int(valor) == int(codigo_produto.codigo):
          return i
      return -1

    if tipo == 'nome':
      for i in range(self.ultima_posicao + 1):
        codigo_produto = self.produtos[i]
        if valor == codigo_produto.nome:
          return i
      return -1
  
  def ordenar(self, tipo):
    if tipo == 'preco':
      produtos_nao_vazios = filter(lambda produto: produto is not None, self.produtos)
      ordenacao = sorted(produtos_nao_vazios, key=lambda produto: produto.preco, reverse=True)
      for i, produto in enumerate(ordenacao):
        print(f'indice {i} - Nome: {produto.nome} - código: {produto.codigo}, quantidade: {produto.quantidade}, preço: {produto.preco}, avaliações: {produto.avaliacao}')
    
    if tipo == 'avaliacao':
      produtos_nao_vazios = filter(lambda produto: produto is not None, self.produtos)
      ordenacao = sorted(produtos_nao_vazios, key=lambda produto: produto.avaliacao, reverse=True)
      for i, produto in enumerate(ordenacao):
        print(f'indice {i} - Nome: {produto.nome} - código: {produto.codigo}, quantidade: {produto.quantidade}, preço: {produto.preco}, avaliações: {produto.avaliacao}')

  def excluir(self, valor, tipo):
    posicao = self.buscar(valor, tipo)
    if posicao == -1:
       return -1
    else:
      for i in range(posicao, self.ultima_posicao):
        self.produtos[i] = self.produtos[i + 1]
      self.ultima_posicao -= 1

  def visualizar(self):
    if self.ultima_posicao == -1:
      print('O vetor está vazio')   
    else:
      for i in range(self.ultima_posicao +1):
        produto = self.produtos[i]
        print('posicao: ', i, ' | nome: ', produto.nome, ' | codigo: ', produto.codigo, ' | quantidade: ', produto.quantidade, ' | preco: ', produto.preco, ' | avaliacao: ', produto.avaliacao) 
        
  def calcular(self): 
    for i in range(self.ultima_posicao +1):
      produto = self.produtos[i]
      total = int(produto.quantidade) * int(produto.preco)
    return total
```
